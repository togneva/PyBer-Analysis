```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color = 'green', marker = 'D', linewidth = 2, label = 'Boston')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Average Fare ($)')
ax.set_xlabel('Month')
ax.grid()
```

## Scatter Plots and Regression Analysis

### MATLAB Method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis,  color = 'red', label = 'Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Month')
plt.ylabel('Fare ($)')
plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
y_axis_larger = []
for y in y_axis:
   y_axis_larger.append(y*3)
```

```python
y_axis_larger
```

```python
plt.scatter(x_axis, y_axis, s = [y*3 for y in y_axis])
```

### Object Oriented Method

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [y*5 for y in y_axis], color = 'skyblue', alpha = 0.8, edgecolor = 'black', linewidth = 2, label = 'Boston')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Month')
ax.set_xlabel('Fare ($)')
ax.set_xlim(0, 50)
ax.invert_yaxis()
```

## Pie Charts

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

### Object Oriented Method

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = [0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
fig, ax = plt.subplots(figsize = (8, 8))
ax.pie(y_axis, 
       labels = x_axis,
       startangle = 90,
       explode = explode_values,
       counterclock = False,
       colors = colors,
       shadow = True,
       autopct = '%.1f%%')
plt.show()
```
